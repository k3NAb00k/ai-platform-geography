# Introduction

Artificial intelligence is increasingly framed as a cornerstone of economic growth, national competitiveness, and technological leadership. As AI industries expand, the geographical distribution of AI jobs carries important implications for regional inequality, urban development, and access to high-skill employment. AI labor markets are often assumed to be geographically concentrated in major urban technology hubs, like Silicon Valley in the United States. In India, cities such as Bengaluru, Hyderabad, and Pune are widely recognized as centers of software development and AI innovation. Urban economic theory suggests that high-skill industries cluster spatially because firms benefit from knowledge spillovers, dense labor markets, and shared infrastructure. However, the rise of digital labor platforms such as global talent intermediaries and remote hiring marketplaces may be altering how AI job opportunities are distributed and made visible. If AI employment continues to cluster in a small number of metropolitan centers, it may deepen spatial inequalities, concentrating opportunity, income, and innovation capacity in already advanced regions.

This project investigates whether platform intermediation reshapes the observed geography of AI job posting in India. Using a combined dataset of 2,129 AI-related job postings scraped from open online sources, I compare the geographic concentration of postings from traditional employers with those mediated by digital talent platforms. Specifically, I ask whether platform-based job postings are more geographically dispersed than non-platform postings, and whether they reduce the apparent dominance of major hubs such as Bengaluru. If platform-mediated hiring disperses job posting across a wider range of cities, it could signal a partial decoupling of AI labor markets from traditional tech hubs. Conversely, if platform postings replicate existing geographic concentration, digital intermediation may reinforce rather than disrupt urban clustering. Understanding whether digital platforms redistribute or reinforce geographic concentration is critical for policymakers, firms, and workers navigating the future of AI-driven economic transformation. By examining employer concentration, geographic clustering, and distributional structure, this study contributes to broader debates about urban agglomeration, digital intermediation, and the changing spatial organization of high-skill work.





To evaluate whether digital platforms reshape the geography of AI hiring, it is necessary to situate this question within existing scholarship on urban agglomeration, labor market structure, and digital intermediation. Research in urban economics has long emphasized the tendency of high-skill industries to cluster in dense metropolitan hubs, where proximity facilitates knowledge spillover, specialized labor pooling, and innovation. At the same time, emerging scholarship on platform capitalism and digital labor markets suggests that intermediated matching systems can reorganize how work is distributed and accessed, potentially weakening geographic constraints. These bodies of literature generate competing expectations: agglomeration theory predicts continued concentration of AI employment in established hubs, while platform theory raises the possibility of spatial dispersion through digitally mediated hiring. The following review outlines these perspectives and clarifies the theoretical framework guiding this study.

# Literature Review
**Agglomeration and the Geography of High-Skill Employment**

A large body of research in urban economics and labor economics demonstrates that high-skill industries have a tendency to cluster geographically due to agglomeration economics, which are the benefits that firms gain when they locate near one another. Rosenthal and Strange (2004) synthesize evidence showing that these spatial concentrations arise from knowledge spillovers, labor market pooling, and input sharing, all of which increase the firm’s productivity in dense metropolitan areas. Similarly, Glaeser and Kerr (2008) found that entrepreneurship and innovation are strongly influenced by local industrial conditions, reinforcing the benefits of spatial clustering. These mechanisms suggest that AI-related employment in India should follow similar concentration patterns in technology hubs such as Bengaluru and Hyderabad, where large labor markets and knowledge networks for AI-related jobs already exist.

Moretti (2010) extends this logic by highlighting how local labor markets shape employment outcomes and wages through multiplier effects and skill complementaries. He argues that when a high-skill job is created in a city, it does not only benefit that worker, but generates local economic activity through housing, transportation, and various services that worker spends their salary on, boosting the economy in a variety of ways. We've already established that high-skill workers are also more productive when they are near other high-skill workers, which leads to firms being willing to pay more in those regions because of this increase in productivity, furthering each workers’ local economic impact.  Following this framework, AI hiring should follow this urban advantage, concentrating opportunities and income in regions that already dominate. If AI employment reflects these agglomeration dynamics, we would expect job postings to be disproportionately clustered in India’s current leading tech centers.

**Technological Change and Labor Market Structure**

Research on technological change further underscores the uneven spatial and occupational consequences of skill based innovation. Autor (2014) argues that technological progress reshapes labor demand in ways that can increase inequality and alter employment structures. Technological change does not affect all workers equally, high-skill workers see more increase productivity,  expanding the scale of what they can do with new changes, in turn making them more valuable workers, whereas middle-skill workers hold positions that are now easier to digitize or outsource, potentiall stunting their wage growth and career stability. Although Autor's analysis is not primarily geographic, it does provide a framework for understanding how AI may intesify the concentration of high-wage employment in regions that are already equipped to support it. This reinforces the expectation that AI job markets could deepen spatial disparities if left to agglomeration forces alone.

**Digital Platforms and Labor Market Intermediation**

In contrast to agglomeration theory, scholarship on digital platforms suggests that technological intermediation can reshape labor market matching processes. Kenney and Zysman (2016) describe digital platforms as a new organizational form that restructures industries by mediating interactions between users, reducing transaction costs, and reorganizing value chains. These platforms don't just sell products, they sit between groups and facilitate interaction between a firm and a worker. By centralizing job listings, standardizing information, and algorithmically matching firms with workers, digital platforms reduce search frictions and insert themselves into traditional hiring processes, thereby reshaping labor market organization.


Agrawal et al. (2019) specifically examine digital digitization in contract labor markets and argue digital technologies can expand market reach and alter contracting structures. By lowering matching costs and enabling remote coordination, platforms may weaken traditional geographic constraints. From this perspective, AI job postings mediated by digital talent platforms could exhibit greater geographic dispersion than postings from traditional employers.

**Remote Work and Spatial Decoupling**

Recent empirical work on remote work further complicates expectations about geographic concentration. Dingel and Neiman (2020) estimate the share of occupations that can be performed remotely and find that remote-capable jobs are disproportionately high-skill and concentrated in knowledge sectors. This suggests that AI roles may be structurally compatible with geographical dispersion. However, Althoff et al. (2022) demonstrates that even remote work remains geographically patterned, as workers and firms maintain ties to major economic hubs. This demonstrates that remote work does not automatically eliminate agglomeration effects; instead, it may reshape them.

# Analysis

# Section 1: Employer Structure and Market Concentration
Before examining geographic patterns, I first assess the overall structure of employer participation in the AI job market. If a small number of firms dominate the postings in this dataset, observed geographic patterns may reflect employer concentration rather than broader labor market dynamics. To evaluate this, I examine the distribution of postings across companies and calculate a Herfindahl–Hirschman Index (HHI) to measure concentration.


```python
df_all['title'].value_counts().head(20)
df_all['location'].value_counts().head(20)
df_all['company'].value_counts().head(20)
```

The dataset contains 2,129 job postings. A substantial share of listings comes from a small number of companies. Turing accounts for 284 postings, representing approximately 13.3% of the entire dataset. When combined with Uplers, the top two firms account for 18.4% of postings—nearly one in five jobs. Notably, both firms are global talent platforms rather than traditional technology companies. **This initially suggests that AI hiring visibility in India may be disproportionately mediated by digital labor intermediaries rather than domestic product firms.**

To further evaluate this hunch, I examine the full distribution of postings across employers.

```python
#Visualize the long-tail structure
company_counts = df_all['company'].value_counts()
plt.figure(figsize=(8,6))
plt.plot(company_counts.values)
plt.yscale('log')
plt.title("Log-Scale Company Distribution")
plt.xlabel("Company Rank")
plt.ylabel("Number of Postings (Log Scale)")
plt.show()
```

This figure plots the number of AI job postings per company, ranked from most to least frequent, using a logarithmic scale on the y-axis. The steep initial decline followed by a gradual flattening indicates a classic long-tail distribution: a small number of firms post many jobs, while the majority of firms post only one or two postings.

Although firms such as Turing appear prominently at the top of the distribution, the rapid decline and extended tail demonstrate that participation in the AI job market is broadly dispersed across many employers. The visual pattern suggests that while platform intermediaries are highly visible, the overall market structure is not monopolized by a small oligopoly.

To quantify this concentration, I calculate the Herfindahl–Hirschman Index (HHI):

```python
#Herfindahl-Hirschman Index (HHI)
company_share = df_all['company'].value_counts(normalize=True)
hhi = (company_share ** 2).sum()
print("HHI:", hhi)
```

# Interpretation
The HHI equals 0.023, which is considered very low. In standard industrial organization benchmarks, values below 0.10 indicate a highly competitive or dispersed market structure. Despite the visibility of leading platform firms, the presence of many employers with small numbers of postings substantially lowers overall concentration.

# Conclusion of Section 1

These results suggest a hybrid market structure. While digital labor platforms account for a notable share of visible postings, the broader AI job market in this dataset remains structurally competitive. Platform firms appear to “punch above their weight” in visibility, but they do not dominate the market in a structural sense. This finding complicates the initial suspicion of platform dominance and motivates further analysis of how platform intermediation may instead shape geographic distribution rather than overall market concentration.

# Section 3: Geographic Concentration

To evaluate whether platform intermediation reshapes geographic patterns, I compare the spatial distribution of job postings between platform and non-platform employers. Agglomeration theory predicts that both groups should exhibit similar hub concentration if geographic clustering reflects underlying labor market fundamentals. However, if digital platforms weaken geographic constraints by lowering search and coordination costs, platform postings should display greater spatial dispersion.

Before comparing distributions, I standardize location values and extract state-level information to reduce inconsistencies in geographic labels.

```python
#Cleaning the location values
df_all['location'] = df_all['location'].replace({
    'KA, IN': 'Karnataka, India',
    'MH, IN': 'Maharashtra, India',
    'TS, IN': 'Telangana, India',
    'TN, IN': 'Tamil Nadu, India',
    'UP, IN': 'Uttar Pradesh, India'
})
#Extracting state
df_all['state'] = df_all['location'].str.split(',').str[1].str.strip()

#Finding geographical distribution (non-platform)
df_all.groupby('is_platform')['location'].value_counts(normalize=True).head(15)
```

# Descriptive Geographic Patterns

The distribution of non-platform postings reveals strong geographic clustering. Approximately 27.7% of non-platform postings are located in Bengaluru, with Pune and Hyderabad forming secondary hubs. This pattern is consistent with established agglomeration dynamics in India’s technology sector.

```python
#Finding geographical distribution (platform)
platform_geo = (
    df_all[df_all['is_platform'] == True]
    ['location']
    .value_counts(normalize=True)
    .head(15)
)

platform_geo
```

In contrast, platform-mediated postings are substantially more geographically dispersed. Bengaluru accounts for only 8.1% of platform postings, and the top 15 platform locations include a wider range of cities, including Tier-2 urban centers such as Jaipur, Lucknow, Coimbatore, and Mysuru

```python
#Preparing to visualize
top_locations = (
    df_all['location']
    .value_counts()
    .head(10)
    .index
)
plot_df = (
    df_all[df_all['location'].isin(top_locations)]
    .groupby(['location', 'is_platform'])
    .size()
    .reset_index(name='count')
)
#Plotting
plt.figure(figsize=(12,6))
sns.barplot(data=plot_df, x='location', y='count', hue='is_platform')

plt.xticks(rotation=45, ha='right')
plt.title("Top Locations: Platform vs Non-Platform AI Jobs")
plt.xlabel("Location")
plt.ylabel("Number of Job Postings")
plt.legend(title="Platform Job")
plt.tight_layout()
plt.show()
```

The side-by-side bar chart further illustrates this divergence: traditional firms are heavily concentrated in Bengaluru, while platform postings are more evenly distributed across multiple locations.

```python
#Comparing top city dominance
non_platform_top = (
    df_all[df_all['is_platform'] == False]
    ['location']
    .value_counts(normalize=True)
    .iloc[0]
)

platform_top = (
    df_all[df_all['is_platform'] == True]
    ['location']
    .value_counts(normalize=True)
    .iloc[0]
)

plt.figure(figsize=(6,5))
plt.bar(['Non-Platform', 'Platform'],
        [non_platform_top, platform_top])

plt.title("Top City Share Comparison")
plt.ylabel("Proportion of Jobs in Top City")
plt.show()
```

# Top-City Dominance Comparison

The difference in hub dominance becomes clearer when comparing the share of postings located in the top city for each group. While over one-quarter of non-platform postings are concentrated in Bengaluru, the top-city share for platform postings is substantially lower. This suggests that platform-mediated hiring reduces the apparent dominance of India’s primary AI hub.

# Geographic Concentration (HHI)

To quantify this difference, I calculate separate Herfindahl–Hirschman Index (HHI) values for platform and non-platform postings.

```python
#Calculate geographic HHI separately
non_platform_share = (
    df_all[df_all['is_platform'] == False]
    ['location']
    .value_counts(normalize=True)
)

platform_share = (
    df_all[df_all['is_platform'] == True]
    ['location']
    .value_counts(normalize=True)
)

hhi_non_platform = (non_platform_share ** 2).sum()
hhi_platform = (platform_share ** 2).sum()

print("HHI Non-Platform:", hhi_non_platform)
print("HHI Platform:", hhi_platform)
#HHI Comparison
plt.figure(figsize=(6,5))
plt.bar(['Non-Platform HHI', 'Platform HHI'],
        [hhi_non_platform, hhi_platform])
plt.title("Geographic Concentration (HHI) Comparison")
plt.ylabel("HHI Value")
plt.show()
```

Non-platform postings exhibit a higher geographic HHI, indicating stronger spatial concentration. Platform postings show a lower HHI, reflecting a flatter distribution across cities. This numerical comparison reinforces the descriptive evidence: traditional employers follow stronger agglomeration patterns, while platform-mediated postings are more geographically dispersed.

```python
#Adding a logistic regression
# Make sure location is a string and not missing
df_reg = df_all.dropna(subset=['location', 'is_platform']).copy()
df_reg['location'] = df_reg['location'].astype(str)

# 0/1 outcome: is the job in Bengaluru?
df_reg['is_bengaluru'] = df_reg['location'].str.contains('Bengaluru', case=False, na=False).astype(int)

# 0/1 predictor: platform job?
df_reg['is_platform'] = df_reg['is_platform'].astype(int)

model = smf.logit("is_bengaluru ~ is_platform", data=df_reg).fit(disp=False)
print(model.summary())
```

# Logistic Regression: Statistical Test of Bengaluru Concentration

To formally test whether platform postings are less likely to be located in Bengaluru, I estimate a logistic regression model predicting whether a job posting is located in Bengaluru as a function of platform status.
